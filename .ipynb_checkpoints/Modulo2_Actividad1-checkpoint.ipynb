**Máquina de Monedas**

Para despachar dinero en efectivo con una máquina automática realice lo siguiente:

a) Escriba una función separanum(valor), que reciba un valor numérico real y separe las cantidades en un vector que contiene: Centavos, unidadesm decenas, centenas, etc.  Los centavos equivalen a usar dos decimales redondeade al entero más cercano.  Desarrolle el algoritmo usando operaciones de cociente y residuo (no use round()).

```python
# Importa librería Numpy
import numpy as np

# Crea función separanum que recibe como argumento un valor y los separa en decimales, unidades, decenas, centenas, etc
def separanum(valor):
    
    # Expresion para realizar el redondeo a dos decimales si es que se ingresan mas de 2 digitos decimales
    centavos = int((valor * 100 + 0.5) % 100)
    
    # Ya obtenidos los decimales, transformamos el valor ingresado a numero entero lo que eliminará los decimales dejando solo el valor entero
    valor_entero = int(valor)
    
    # Convertir el valor entero en un array de dígitos, convirtiendo el entero en una cadena y luego iterando por cada caracter de la cadena
    # los transformamos a numeros enteros, finalmente con metodo np.array() obtenemos un array con los numeros separados
    digitos = np.array([int(d) for d in str(valor_entero)])
    
    # Invertimos el array digitos
    digitos = digitos[::-1]
    
    # Creamos una expresion que contiene el valor de los centavos en una lista y concatenamos con el array digitos convertidos en lista
    cantidades = [centavos] + digitos.tolist()

    # Devuelve el resultado de la variable cantidades
    return cantidades


# Líneas que inicializan el programa solicitando el ingreso de un valor al usuario y pasando el valor como argumento a la función.
valor = float(input('Ingrese un monto por favor: '))
cantidades = separanum(valor)

# Se imprime el resultado de cantidades
print(cantidades)
```

b) Realice una función cuantasmonedas(centavos) que reciba la cantidad de centevaos a entregar con monedas de 1, 5, 10, 25, 50 centavos usando la menor cantidad de monedas.s.

```python
# Importamos librería Numpy
import numpy as np

# Creamos función cuantasmonedas que recibe como argumento un valor en centavos
def cuantasmonedas(centavos):
    
    # Creamos variable 'valores_monedas' usando metodo np.array que establece el valor de las monedas en un array
    valores_monedas = np.array([50, 25, 10, 5, 1])
    
    # Creamos variable 'cantidad_monedas' con un array de ceros con la misma cantidad de elementos que el array valores_monedas
    cantidad_monedas = np.zeros_like(valores_monedas)
    
    # Crea bucle for que calcula la cantidad de monedas mínimas para completar el valor
    # El bucle itera sobre los pares (indice, valor_moneda) creado por la funcion enumerate,
    # correspondiendo al ('i' al indice, y 'valor' al valor de moneda)
    for i, valor in enumerate(valores_monedas):

        # Expresión que calcula la cantidad de monedas por cada indice [i].  El calculo se hace usando operador '//' que devuelve
        # la división de entero
        cantidad_monedas[i] = centavos // valor
        # Actualiza el valor de la variable centavos mediante la operacion de modulo entre centavos y el valor de la moneda iterada
        centavos %= valor
    
    # Crear un diccionario con los resultados
    resultado = {
        '50 centavos': cantidad_monedas[0],
        '25 centavos': cantidad_monedas[1],
        '10 centavos': cantidad_monedas[2],
        '5 centavos': cantidad_monedas[3],
        '1 centavo': cantidad_monedas[4]
    }
    
    return resultado

# Lineas que incializan el programa
centavos = int(input('Ingrese el monto en centavos para calcular el monto en monedas: '))

# Crea variable resultado que incializa la funcion cuantasmonedas pasando como argumento la variable centavos
resultado = cuantasmonedas(centavos)

# Imprime respuesta
print(f"Para {centavos} centavos se necesitan:")
for moneda, cantidad in resultado.items():
    print(f"{cantidad} monedas de {moneda}")
```

3. Realice un programa que permita crear un calendario de un mes cualquiera, conociendo el día de la semana que inicia el mes y el numero de días que tiene el mes..


```python
# Importa libraria Numpy
import numpy as np

# Crea función crear_calendario que usa métodos de Numpy para su construcción
def crear_calendario(dia_inicio, num_dias):
    
    # Variable dias de la semana
    dias_semana = np.array(['Domingo', 'Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado'])

    # Código para imprimir encabezado del calendario
    print("     Calendario")
    print("--------------------")
    print(" D  L  M  X  J  V  S")
    print("--------------------")
    
    # Crea bloque de manejo de excepciones en expresión que obtiene índice de día de inicio
    try:
        indice_dia_inicio = np.where(dias_semana == dia_inicio)[0][0]

    # Captura error y lo maneja
    except ValueError:
        print("Entrada inválida. Intente de nuevo.")
        return

    # Crea variable de inicialización del contador de día
    contador_dia = 1

    # Crea variable filas_calendario como lista vacía
    filas_calendario = []

    # Establece variable semana para construir filas del calendario
    semana = 0

    # Crea bucle while que se ejecuta para construir todas las filas del calendario mientras el valor del día sea menor o igual a la 
    # cantidad de días del mes ingresado por el usuario
    while contador_dia <= num_dias:
        fila = []

        # Crea bucle for que se ejecuta por cada día de la semana (7 veces)
        for dia in range(7):
            # Condicional que evalua el dia de inicio del mes y agrega '0' en caso de que el primer dia del mes no sea domingo
            if semana == 0 and dia < indice_dia_inicio:
                fila.append(" 0 ")

            # Bloque que evalua si el contador_dia es menor al numero de días y va agregando el valor hasta que se cumpla el total de días del mes
            # luego incrementa en 1 unidad el valor del contador día
            elif contador_dia <= num_dias:
                fila.append(f"{contador_dia:2} ")
                contador_dia += 1

            # Cuando se cumplen las condiciones anteriores y si aún quedan días de las semana vacíos en la última fila, los rellena con 0
            else:
                fila.append(" 0 ")

        # Agreag la fila construida a la lista filas_calendario e incrementa (+1) el valor de la variable contador semana
        filas_calendario.append(fila)
        semana += 1

    # Bucle for para imprimir las filas del calendario
    for fila in filas_calendario:
        print("".join(fila))

# Bucle while para asegurar el correcto ingreso de el día de inicio y convierte la primera letra de el ingreso a Mayuscula
while True:
    dia_inicio = input("Ingrese el día de la semana en que inicia el mes (Domingo, Lunes, Martes, Miércoles, Jueves, Viernes, Sábado): ")
    dia_inicio = dia_inicio.capitalize()

    # Bloque que evalúa que lo ingresado por el usuario sea válido buscando el ingreso en la lista de días de la semana
    if dia_inicio in ['Domingo', 'Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado']:
        # Sale del bucle
        break

    # En caso de que no se ingrese de manera correcta el día de la semana, imprime mensaje de advertencia de error y vuelve
    # al inicio del bucle
    else:
        print("Entrada inválida para el día de la semana. Intente de nuevo.")

# Bucle while con manejo de excepciones para capturar ingresos incorrectos
while True:

    try:

        # Pide ingreso de la cantidad de días del mes e incluye bloque condicional para asegurar un ingreso real de días
        num_dias = int(input("Ingrese el número de días que tiene el mes: "))
        if 28 <= num_dias <= 31:

            # Si se ingresa correctamente la cantidad de días, sale del bucle
            break

        # Imprime el motivo del error si no se ha realizado correctamente
        else:
            print("El número de días debe estar entre 28 y 31. Intente de nuevo.")
    
    # Captura la excepción para evitar la caída del programa
    except ValueError:
        print("Entrada inválida para el número de días. Intente de nuevo.")

# Inicia el programa tomando los datos ingresados por el usuario y pasandolos como argumentos a la función
crear_calendario(dia_inicio, num_dias)

```
