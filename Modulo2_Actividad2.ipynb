1. Crea un programa donde el usuario debe ingresar las ventas en un periodo de años y aparezca en pantalla el conjunto de datos de ventas por año y al aplicarle un descuento del 15%.  (Debe aparecer el antes y el despues).s).

```python
# Importa librería Pandas para uso y construcción de Data Frames
import pandas as pd

# Crea función que solicita los datos para la construcción del Data Frame
def cuadro_ventas_por_anio():
    # Solicitar al usuario el número de años y las ventas por año
    num_anios = int(input('Ingrese el número de años: '))
    
    # Inicializa diccionario vacío para el almacenamiento de la información (clave)anio: (valor)ventas
    ventas_por_anio = {}

    # Bucle for que itera sobre el rango de años ingresados por usuario
    for anio in range(1, num_anios + 1):
        
        # Solicita el ingreso de las ventas por año en cada iteración
        ventas = float(input(f'Ingrese las ventas para el año {anio}: '))
        # Agrega al diccionario la (clave:valor) anio:ventas en cada iteración
        ventas_por_anio[f'{anio}'] = ventas
    
    # Crea un DataFrame de dos columnas en que los datos se rescatan del resultado de (ventas_por_anio.items())
    df_ventas = pd.DataFrame(ventas_por_anio.items(), columns=['Año', 'Ventas'])
    
    # Crea dos nuevas columnas de nombre 'Descuento' y 'Total (Desc. Incluido)' en df_ventas

    # Columna 'Descuento' se establece como valor fijo el % de descuento
    df_ventas['Descuento'] = '15%'
    # Columna 'Total (Desc. Incluido)' toma los datos de la columna 'Ventas' del DataFrame y multiplica por 0.85 para obtener el total descuento aplicado
    df_ventas['Total (Desc. Incluido)'] = df_ventas['Ventas'] * 0.85
        
    
    # Imprime un enunciado y el df_ventas
    print('\nResumen de ventas:')
    print(df_ventas)

# Lineas para iniciar el programa        
if __name__ == '__main__':
    cuadro_ventas_por_anio()
```

2. Construya un programa que guarde un diccionario que contenga las notas del examen del curso de Análisis de Datos del presente año, y debe devolver la nota mínima, la nota máxima, la nota promedio y su desviación estándarar.

```python
# Importa librería pandas
import pandas as pd

def programa_notas_curso():
    # Inicializa diccionario para almacenamiento de los datos
    notas_curso_analisis_datos = {}
    
    # Crea lista de alumnos para automatización en el ingreso de las notas
    estudiantes = ['Luca', 'Facundo', 'Magdalena', 'Olimpia', 'Trinidad', 'Marina', 'Isabella', 'Matteo']
    
    
    
         
    # Bucle for para iterar sobre lista estudiantes y solicitar el ingreso de las notas del curso
    for estudiante in estudiantes:
    
        # Bucle while para asegurar el correcto ingreso de los datos
        while True:
            # Bloque para manejo de excepciones
            try:
                nota = float(input(f'Ingrese la nota para {estudiante}: '))
                # Bloque condicional que controla el ingreso de notas en el rango 0 a 7
                if 0 <= nota <= 7:
                    # Agrega los datos ingresados por el usuario al diccionario de notas del curso 'estudiante:nota'
                    notas_curso_analisis_datos[f'{estudiante}'] = nota
                    # Sale del bucle si se ha realizado el ingreso correctamente
                    break
                # Si la nota no está en el rango correcto, imprime mensaje informativo y vuelve a solicitar el ingreso
                else:
                    print('--' * 15)
                    print('La nota debe estar en el rango de 0 a 7. Intente nuevamente')
                    print('--' * 15)
            # Instrucción que captura la excepción ValueError si se ingresa un dato que no sea del tipo numérico
            except ValueError:
                print('--' * 15)
                print('Debe ingresar un valor númerico')
                print('--' * 15)
    
    df_notasCurso = pd.DataFrame(notas_curso_analisis_datos.items(), columns=['Estudiante', 'Nota'])
    return df_notasCurso


if __name__ == '__main__':
   df_calificaciones_curso = programa_notas_curso()
    
print('')
print('--' * 22)
print('--' * 22)
print('Cuadro de Notas Curso Análisis de Datos')
print('--' * 22)
print('--' * 22)
print(df_calificaciones_curso)
print('--' * 22)
        
# Calculo de estadísticas Nota Máxima, Nota Mínima, Promedio de Notas Curso y Desviación Estándar de notas
        
nota_maxima = df_calificaciones_curso['Nota'].max()
nota_minima = df_calificaciones_curso['Nota'].min()
nota_promedio = df_calificaciones_curso['Nota'].mean()
desviacion_standar_notas = df_calificaciones_curso['Nota'].std()
        
print(f'La nota más alta del curso es {nota_maxima}')
print(f'La nota más baja del curso es {nota_minima}')
print(f'El promedio de notas del curso es {nota_promedio}')
print(f'La desviación estándar de las notas del curso es {desviacion_standar_notas}')
    
```

4. Construya un programa que pregunte y guarde un Data Frame usando los datos de la tabla adjunta (No se trata de crear 'a mano' el Data Frame)

<img src="tablaEjercicio2_4.jpg">

```python
import pandas as pd

data = {}

meses = []
ventas = []
gastos = []
meses_anio = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
salir = False

while not salir:
    print('---' * 20)
    while True:
        mes = input('Ingrese el mes: ').capitalize()
        if mes in meses_anio:
            meses.append(mes)
            break
        else:
            print('Ese mes no existe')
        

    while True:
        try:
            venta = float(input('Ingrese el monto de las ventas: '))
            if venta >= 0:
                ventas.append(venta)
                break
            else:
                print('Debe ingresar un valor numérico mayor a 0')
        except ValueError:
            print('Error: Debe ingresar un valor numérico mayor a 0')

    while True:
        try:
            gasto = float(input('Ingrese el monto de los gastos: '))
            if gasto >= 0:
                gastos.append(gasto)
                break
            else:
                print('Debe ingresar un valor numérico mayor a 0')
        except ValueError:
            print('Error: Debe ingresar un valor numérico mayor a 0')
    
    print('---' * 20)
    exit = input('¿Desea seguir ingresando datos? (s/n): ')
    if exit != 's':
        break


data['Mes'] = meses
data['Ventas'] = ventas
data['Gastos'] = gastos

df = pd.DataFrame(data)

df
```
